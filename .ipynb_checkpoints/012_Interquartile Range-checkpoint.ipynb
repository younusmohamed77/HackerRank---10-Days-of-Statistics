```python
def find_median(data):
    n = len(data)
    sorted_data = sorted(data)
    midpoint = n // 2
    
    if n % 2 == 1:
        # Odd length
        return sorted_data[midpoint]
    else:
        # Even length
        return (sorted_data[midpoint - 1] + sorted_data[midpoint]) / 2

def quartiles(data):
    n = len(data)
    sorted_data = sorted(data)
    
    # Find Q2 (the median)
    Q2 = find_median(sorted_data)
    
    # Split the data into two halves for Q1 and Q3
    if n % 2 == 0:
        lower_half = sorted_data[:n // 2]
        upper_half = sorted_data[n // 2:]
    else:
        lower_half = sorted_data[:n // 2]
        upper_half = sorted_data[n // 2 + 1:]
    
    # Find Q1 and Q3
    Q1 = find_median(lower_half)
    Q3 = find_median(upper_half)
    
    out_list = []
    out_list.append(Q1)
    out_list.append(Q2)
    out_list.append(Q3)
    
    return out_list

def interQuartile(values, freqs):
    S = []
    
    for i in freqs:
        for j in range(i):
            S.append(values[i])
            
    quartile_values = quartiles(S)
    
    iqr = quartile_values[2] - quartile_values[0]
    
    print('{:.1}'.format(iqr))
```

```python
def find_median(data):
    n = len(data)
    sorted_data = sorted(data)
    midpoint = n // 2
    
    if n % 2 == 1:
        # Odd length
        return sorted_data[midpoint]
    else:
        # Even length
        return (sorted_data[midpoint - 1] + sorted_data[midpoint]) / 2

def quartiles(data):
    n = len(data)
    sorted_data = sorted(data)
    
    # Find Q2 (the median)
    Q2 = find_median(sorted_data)
    
    # Split the data into two halves for Q1 and Q3
    if n % 2 == 0:
        lower_half = sorted_data[:n // 2]
        upper_half = sorted_data[n // 2:]
    else:
        lower_half = sorted_data[:n // 2]
        upper_half = sorted_data[n // 2 + 1:]
    
    # Find Q1 and Q3
    Q1 = find_median(lower_half)
    Q3 = find_median(upper_half)
    
    return [Q1, Q2, Q3]

def interQuartile(values, freqs):
    S = []
    
    for value, freq in zip(values, freqs):
        S.extend([value] * freq)
    
    quartile_values = quartiles(S)
    
    iqr = quartile_values[2] - quartile_values[0]
    
    print('{:.1f}'.format(iqr))
```
